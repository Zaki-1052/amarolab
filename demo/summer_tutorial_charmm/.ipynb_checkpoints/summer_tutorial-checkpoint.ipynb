# **Molecular Dynamics with OpenMM**

[OpenMM](https://openmm.org/) is a MD engine that is open-source, highly customizable, and optimized for simulation on GPUs. The Python wrapper makes setting up simulations fairly easy, and allows us to do more complicated things later (like simulating membrane proteins). The following tutorial covers all necessary steps for setting up a simple simulation of a solvated protein system using the CHARMM forcefield.

For this tutorial, we will be using **ubiquitin** (PDB:1UBQ) as a model system.

```python
# Import all the packages we need for the tutorial
from openmm import *
from openmm.app import *
import openmm.unit as unit
import nglview as nv
import MDAnalysis as mda
from MDAnalysis.transformations import unwrap, center_in_box, wrap
from sys import stdout
```

```python
# Visualize PDB structure of ubiquitin
view = nv.show_pdbid("1ubq")
view
```

## Parameters

### 1. Define system files and forcefield parameters

Our simulation requires several files: a **structure file**, a **topology file**, and **parameter** files. The **structure** contains all the information about the positions and types of the atoms in your system (usually a file from the PDB that has been prepared to include waters, hydrogens, ions, etc.). The **topology** file provides information about how those atoms are connected with bonds. The **parameter** file contains the parameters to use in the potential energy function. The structure is usually a .pdb file for most MD simulations, while the parameter and topology files depend on the force field being used. For CHARMM, the .psf file contains the topology information, while the parameter information is contained in multiple files: .rtf, .prm, etc. CHARMM-GUI will give you the necessary files in /toppar. For this tutorial, we have a prepared system ready to go.

```python
# Load our .pdb and .psf filenames
pdb_filename = "step3_pbcsetup.pdb" # contains positions of atoms
psf_filename = "step3_pbcsetup.psf" # contains list of atoms and how they are connected
```

```python
# Visualize prepared system
view = nv.show_file(pdb_filename)
view.clear_representations()
view.add_cartoon(selection="protein", color="dodgerblue")
view.add_representation("licorice", selection="water")
view.add_representation("spacefill", selection="SOD or CLA", radius=1)
view
```

### 2. Define conditions and variables

Next, we define the conditions for the simulation. How much time between steps where we evaluate the equations of motion? Typically this is 2 femtoseconds, but can be increased up to 4 fs with some clever physical tricks. What temperature do we want to run the simulation at? This depends on what experimental conditions you are trying to replicate, if any. Some of the experimental conditions, like ionic strength and pH, are done at the parameterization stage. Additionally, we need to set the size of the simulation box, which is defined by the box vectors. 

```python
#Temperature for simulation (thermostat will keep it stable at this temp)
temperature = 298.15*unit.kelvin

# Total simulation time = num_steps * time_step
# 0.002 ps * 100000 = 200 ps
num_steps = 100000
time_step = 0.002 * unit.picoseconds

# Box vectors (check step3_pbcsetup.str if not in the pdb)
a = 64 * unit.angstrom
b = 64 * unit.angstrom
c = 64 * unit.angstrom
alpha = 90.0 * unit.degrees
beta  = 90.0 * unit.degrees
gamma = 90.0 * unit.degrees
```

### 3. Set up simulation reporters

If we want to analyze our simulation, we need to save structural snapshots as we let the system evolve. Saving every step would be way too much data, but we need to make sure we are capturing relevant motions. Choose this number based on the timescale of relevant motions. In this case, we will save every 2 picoseconds to a dcd file by using a DCDReporter (a DCD file only saves positions, so it must be loaded with a topology file for visualization later). It is also useful to output some information directly to the console as the simulation progresses, so we can see how it's doing. We will add a StateDataReporter that will report the number of steps completed, the potential energy, and the system temperature.

```python
# Information for DCD reporter
trajectory_filename = "trajectory.dcd"
steps_per_trajectory_update = 1000

# Information for stdout reporter
steps_per_energy_update = 1000
```

## Simulation

### 1. Load files into OpenMM objects
Now, we want to take all the variables and files we have defined and turn them into the objects that OpenMM needs in order to compute forces and propagate dynamics. First, we turn the .pdb, .psf files into OpenMM `PDBFile` and `CharmmPsfFile` objects, which allows OpenMM to store and read the data in these files. We also need to update OpenMM's PsfFile with the box vectors that we defined above, and load the parameter files into an OpenMM `CharmmParameterSet`. 

```python
# Turn these files into OpenMM objects
mypdb = PDBFile(pdb_filename)
mypsf = CharmmPsfFile(psf_filename)

# Update psf with box vectors
mypsf.setBox(a, b, c, alpha, beta, gamma)

# Load the CHARMM parameter filenames (if using CHARMM-GUI for setup, these are in charmm-gui-000000000/toppar/)
# For this simple example, we only need the protein and water params.
myparams = CharmmParameterSet("par_all36m_prot.prm", "top_all36_prot.rtf", "toppar_water_ions.str") # contains parameters for simulation
```

### 2. Create system and integrator

The OpenMM `System` contains all the information about how the forces are computed. At a minimum, it needs the parameter set, which contains all the information it needs to compute forces from the potential energy function. However, we can also include some changes to how the forces are computed in order to make our simulation faster and more stable. Specifically, we can cut off the pairwise nonbonded force calculation past a certain distance for each atom, and then use a long-range electrostatic method (in this case PME) past that distance. You can read more about these methods [here](https://docs.openmm.org/latest/userguide/theory/02_standard_forces.html#nonbondedforce).

The `Integrator` controls how positions and velocities are updated based on the forces. Because we want to keep the temperature constant, we will use a Langevin integrator, which simulates a system in contact with a heat bath by adding stochastic elements to Newtonian mechanics. You can read more about this integrator, as well as other options, [here](https://docs.openmm.org/latest/userguide/theory/04_integrators.html#langevinintegator).

```python
# Create an OpenMM system
system = mypsf.createSystem(
            myparams, # pass the parameters to the system
            nonbondedMethod=PME, # Use particle mesh Ewald to compute long-range electrostatics
            nonbondedCutoff=1.0*unit.nanometer, # Cutoff past which to use long-range electrostatics (PME)
            constraints=HBonds # Constrain bonds to hydrogen to make more stable
)
```

```python
# Create integrator
integrator = LangevinIntegrator(temperature, 1/unit.picosecond, time_step) #1/picosecond is the friction coefficient
```

### 3. Create simulation
The `Simulation` object synthesizes everything we have created so far. It combines the positions from the pdb, the topology from the psf, the `System` with the parameter files, and the `Integrator`. Now, the simulation knows where every atom starts, how they are connected, how to compute the forces, and how to update the positions and velocities.

The simulation also contains all the reporters that we set up earlier, including the trajectory writer that will enable us to analyze our simulation.

```python
# Create an OpenMM simulation object, which combines the positions and topology, the system, and the integrator.
simulation = Simulation(mypsf.topology, system, integrator)

# Update the simulation with the positions of all the atoms
simulation.context.setPositions(mypdb.positions)
```

```python
# Add all our reporters (trajectory and stdout)
simulation.reporters.append(DCDReporter(trajectory_filename, steps_per_trajectory_update))
simulation.reporters.append(StateDataReporter(stdout, steps_per_energy_update, step=True, potentialEnergy=True, temperature=True))
```

Before we can run our simulation, we want to perform an energy minimization. This searches for a local minimum in the potential energy and returns the corresponding set of positions, which helps to resolve any steric clashes or unphysical geometries that might cause our simulation to crash. 

Next, we want to set the initial velocities to match the temperature we want to simulate at. To do this, velocities are randonly assigned to all particles according to the [Maxwell-Boltzmann distribution](https://en.wikipedia.org/wiki/Maxwell%E2%80%93Boltzmann_distribution).

```python
# Minimize energy
simulation.minimizeEnergy(maxIterations=500)

# Set the initial velocities 
simulation.context.setVelocitiesToTemperature(temperature)
```

### 4. Run simulation

```python
# Now we can run it!
simulation.step(num_steps)
```

```python
# Image protein in center of box for visualization 
u = mda.Universe(psf_filename, trajectory_filename)

protein = u.select_atoms("protein")
nonprotein = u.select_atoms("not protein")

u.trajectory.add_transformations(
    unwrap(protein),
    center_in_box(protein, center="geometry", wrap=False),
    wrap(nonprotein, compound="residues")
)
```

```python
# Visualize trajectory
view = nv.show_mdanalysis(u)
view.clear_representations()
view.add_cartoon(selection="protein", color="dodgerblue")
view.add_point(selection="TIP3 and .OH2", radius=0.1)
view.add_ball_and_stick(selection="SOD or CLA", radius=0.35)
view.add_unitcell()
view.center("protein")
view
```
